```python
## resize an image
img_resize= cv2.resize(img,(56,56))
cv2_imshow(img_resize)
```

```python
img_resize.shape

half_img=cv2.resize(img,(img.shape[1] //2,img.shape[0] //2))
cv2_imshow(half_img)
print(img.shape)
print(half_img.shape)
```

```python
#image flip we can use flip to do data augmantation
img_flip = cv2.flip(img,-1)
## 0 flip in vrtcal axis  1 reverse in htal -1 do both
cv2_imshow(img_flip)
```

```python
#image crop
img_crop= img[0:100,0:256] ## first on is H seconfd W
cv2_imshow(img)
cv2_imshow(img_crop)
```

```python
#save image
cv2.imwrite('potato_l.png',img_crop)
```

```python
#drawing shape
new_image= np.zeros((512,512,3))

#recatnagle
cv2.rectangle(new_image,pt1=(50,10),pt2=(400,200),color=(255,0,0),thickness=3)
cv2_imshow(new_image)
```

```python
#circle
cv2.circle(new_image, center=(100,400),radius=50,color=(0,0,255),thickness=-1)
cv2_imshow(new_image)
```

```python
#line
cv2.line(new_image,pt1=(0,0),pt2=(512,512),color=(0,255,0),thickness=2)
# cv2_imshow(new_image)
#text
cv2.putText(new_image,org=(100,100),fontScale=4,color=(0,255,255),thickness=2,lineType=cv2.LINE_4,text="Helloi",fontFace=1)
cv2_imshow(new_image)
```
