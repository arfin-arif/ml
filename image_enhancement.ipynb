```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

#### Background Removel tricks

```python
img = cv2.imread("/content/k-short.png")

img  = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
copy_image=img.copy()
plt.imshow(img)
```

```python
lower=np.array([0,0,0])
higher=np.array([250,250,250])
mask = cv2.inRange(img,lower,higher)
```

```python
mask.shape
```

```python
plt.imshow(mask,'gray')
```

```python
cont,_=cv2.findContours(mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
```

```python
cont_img=cv2.drawContours(img,cont,-1,255,3)
plt.imshow(cont_img)
```

```python
c = max(cont,key=cv2.contourArea)
# c
```

```python
x,y,w,h= cv2.boundingRect(c)
```

```python
cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,3),2)
plt.imshow(img)
```

```python
cropped_img= copy_image[y:y+h,x:x+w]
plt.imshow(cropped_img)
```
